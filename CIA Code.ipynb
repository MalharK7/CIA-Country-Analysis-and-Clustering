```python
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.metrics import silhouette_score

def cluster_countries(feature1, feature2, data):

    selected_features = data[[feature1, feature2]]


    best_model = None
    best_algorithm = None
    best_silhouette_score = -1

    kmeans = KMeans(n_clusters=3, n_init=10, random_state=42)
    kmeans_clusters = kmeans.fit_predict(selected_features)
    kmeans_silhouette_score = silhouette_score(selected_features, kmeans_clusters)


    if kmeans_silhouette_score > best_silhouette_score:
        best_model = kmeans
        best_algorithm = "KMeans"
        best_silhouette_score = kmeans_silhouette_score


    agglomerative = AgglomerativeClustering(n_clusters=3)
    agglomerative_clusters = agglomerative.fit_predict(selected_features)
    agglomerative_silhouette_score = silhouette_score(selected_features, agglomerative_clusters)


    if agglomerative_silhouette_score > best_silhouette_score:
        best_model = agglomerative
        best_algorithm = "Agglomerative Clustering"
        best_silhouette_score = agglomerative_silhouette_score


    return best_model, best_algorithm, best_silhouette_score


```

```python
feature1 = "GDP ($ per capita)"
feature2 = "Phones (per 1000)"
best_model, algorithm_used, silhouette_score = cluster_countries(feature1, feature2, df)


print("Algorithm used:", algorithm_used)


print("Silhouette score:", silhouette_score)


clusters = best_model.labels_


print("Cluster Assignments:")
print(clusters)
```
